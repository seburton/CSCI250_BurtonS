# Data noise

Data provided by physical sensors are often corrupted by noise. 

Two common types of noise are 
* **random**, i.e. continuous small fluctuations, and
* **spiky**, i.e. occasional large fluctuations.

# Data denoising

We can denoise data using digital filtering:
* **mean filters** - replace a value of the time series with an average of the neighbor values within a specified window. 

* **median filters** - replace a value of the time series with the median of the neighbor values within a specified window. 

Consider the time series defined below representing a damped oscillatory signal, 

$$
a(t) = e^{-kt} \sin(2 \pi f t)
$$

corrupted by uniform random noise and lots of large spikes. 

Your task is to filter the data using mean and median filters (use a window equal to 1/4 of the signal period).

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import scipy as sp

import numpy.random as rng
rng.seed(1001)
```

```python
T = 10                     # [s]  total time
F = 0.01                   # [Hz] sampling rate

nt = int(T/F)              # samples
f  = 0.5                   # signal frequency
k  = 0.2                   # signal damping 

t = np.linspace(0,10,nt)   # time axis
```

```python
s = np.sin(2*pi*f*t) * np.exp(-k*t)

# start with the signal
a = s.copy()

# add white noise
wn = rng.uniform(-0.2,+0.2, nt)
a += wn

# add spiky noise
ns = 51
sn = np.zeros(nt)
sn[ rng.randint(1,nt,ns) ] = rng.uniform(-5.0,+5.0,ns)
a += sn
```

```python
plt.figure(figsize=(15,5))

plt.plot(t,a,'k')
plt.xlabel('t') 
plt.ylabel('a') 
plt.ylim(-5.1,+5.1);
```

```python
#import the functions I made myself
import My_Funcs as myfun
```

```python
#make datasets for filtered a

#using my functions
medFilt = myfun.myMedianFilter(a,10) ###### note arbitrary window sizes as second argument
meanFilt = myfun.myMeanFilter(a,10)  #####################################################

#and using scipy
spMedFilt = sp.ndimage.median_filter(a,10)

spMeanFilt = []
for i in range(int(len(a)/10)):
    indexArr = np.zeros_like(a)
    indexArr[10*i : 10*i+10] = 1
    for i in range(10):
        spMeanFilt.append(sp.ndimage.mean(a,indexArr))
```

```python
#plot!
fig, (ax1, ax2, ax3) = plt.subplots(3,1)

ax2.plot(t, medFilt, 'b', label='My median filter', linewidth=0.5)
ax3.plot(t, meanFilt, 'b', label="My mean filter", linewidth=0.5)
ax1.plot(t, a, 'k', label='original data', linewidth=0.5)
ax2.plot(t, spMedFilt, 'r', label='SciPy median filter', linewidth=0.5)
ax3.plot(t, spMeanFilt, 'r',label='SciPy mean filter', linewidth=0.5)

ax1.legend(loc='upper right')
ax2.legend()
ax3.legend()

ax1.set_title("Noise filters: SciPy vs student")
ax3.set_xlabel('t')
ax1.set_ylabel('a')
ax2.set_ylabel('a')
ax3.set_ylabel('a')
```

The median filter is better at filtering spiky noise than the mean filter. For areas with no spikes, the mean filter is better at filtering white noise. 

The period of the sine curve is preserved in both filters. When the window is much bigger the filters become less effective, particularly the mean filter. However, when the window is too small, the noise is not captured in the filter so it also doesn't work very well.
