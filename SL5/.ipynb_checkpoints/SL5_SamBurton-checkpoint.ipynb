# Game of Life

The game is a [cellular automaton](https://en.wikipedia.org/wiki/Cellular_automaton) based on the following rules:
1. Any live cell with fewer than two live neighbors dies, as if by underpopulation.
* Any live cell with two or three live neighbors **lives** on to the next generation.
* Any live cell with more than three live neighbors dies, as if by overpopulation.
* Any dead cell with exactly three live neighbors becomes a **live** cell, as if by reproduction.

        * add the glider to the field only at 10% of the time steps selected randomly
        
    and puts into a queue at every generation its index and a flag
        * the boolean flag specifies if at least 25 cells are alive in its field at that time

    * The **consumer** receives the information from the queue and
        * inserts a glider into the field at a random position every time the received flag is `True`.
        * Once a glider is inserted, the consumer does not insert another one for at least 5 generations
    
    and advances using the GoL logic, like the producer.
    
Thus, the producer and the consumer fields evolve by the same rules. 

They also exchange information (from the producer to the consumer) at specific times according to the rules specified before (i.e. the flag is `True` at all generations when the producer field has 25 live cells).

3. Launch the producer and the consumer in separate threads
    * use a queue to pass information from one thread to another
    * start both threads at the same time (i.e. in very close succession)

4. Plot side-by-side an animation of the two fields, thus visualizing the moments when information is exchanged  between the two universes (from the producer to the consumer).

# Glider

Is a specific shape we will insert into the GoL field. This shape has properties that make the animation interesting, i.e. the shape moves as it changes over generations.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rng

import matplotlib.animation as ani
from IPython.display import HTML
import queue
import time
import threading

plt.set_cmap('Blues')
```

```python
def glider(A, iR,iC):
    A[iR  ,iC-2] += 1
    A[iR  ,iC-1] += 1
    A[iR  ,iC  ] += 1
    A[iR-1,iC  ] += 1
    A[iR-2,iC-1] += 1
```

```python
nR,nC = 8,8

A = np.zeros( [nR,nC], dtype=int)

glider(A, 4,4)

plt.imshow(A)
```

# Animation

Upon completion of the GoL simulation you should have a 3D `numpy` array called `MOVIE` with  $nR$ rows, $nC$ columns, and nSteps slices. 

The following demonstrates how to animate a 3D array - in this case, we will simply add random numbers at various steps. 

```python
nR = 31
nC = 31
nT = 101

MOVIE = np.zeros( [nR,nC,nT], dtype=int)

for iT in range(nT-1):
    iR = rng.randint(1,nR-1)
    iC = rng.randint(1,nC-1)
    MOVIE[iR,iC,iT] = 1
```

```python
golFig = plt.figure(figsize=(5,5)); # create figure

golPlot = plt.imshow(MOVIE[:,:,0]);   # make a plot
plt.axis('off');
```

```python
# define animation function
def golAnimate(i, MOVIE, golP):
    golP.set_data( MOVIE[:,:,i] )
    plt.axis('off')
```

```python
# create inline animation
golMovie = ani.FuncAnimation(fig    = golFig, 
                             func   = golAnimate, 
                             frames = nT,
                             fargs  = (MOVIE,golPlot)
                             );

HTML(golMovie.to_jshtml())       # Java script version (don't use on RPi)
# HTML(rawkMovie.to_html5_video()) # HTML 5 (use on RPi)
```

```python
# this is a function to check all neighbors of a cell and return an array with the number of alive neighbors at each position
def checkNeighbors(Arr):
    nR = Arr.shape[0]
    nC = Arr.shape[1]
    newArr = np.zeros( [nR, nC])
    
    Arr = np.pad(Arr, 1)
    
    #top left
    newArr = newArr + Arr[0:-2, 0:-2]
    #top
    newArr = newArr + Arr[0:-2, 1:-1]
    #top right
    newArr = newArr + Arr[0:-2, 2:]
    #right
    newArr = newArr + Arr[1:-1, 2:]
    #bottom right
    newArr = newArr + Arr[2:, 2:]
    #bottom
    newArr = newArr + Arr[2:, 1:-1]
    #bottom left
    newArr = newArr + Arr[2:, 0:-2]
    #left
    newArr = newArr + Arr[1:-1, 0:-2]
    
    return newArr
```

```python
#This is a function to use the rules to generate the next frame in the movie
def nextGen(Arr, neighbors):
    
    #make arrays of which cells should be alive
    #Any live cell with two or three live neighbors lives on to the next generation.
    #Any live cell with more than three live neighbors dies, as if by overpopulation.
    first = np.where(Arr == 1, Arr, 0) * (np.where(neighbors == 2, neighbors, 0) + np.where(neighbors == 3, neighbors, 0))
    
    second = np.where(Arr == 1, Arr, 0) * np.where(neighbors == 3, neighbors, 0)
    
    #Any dead cell with exactly three live neighbors becomes a live cell, as if by reproduction.
    third = (np.invert(np.where(Arr == 0, Arr, 1))+2) * np.where(neighbors == 3, neighbors, 0)
    
    #sum the condition arrays which note live cells
    newArr = first + second + third
    newArr = np.where(newArr == 0, newArr, 1)
    
    return newArr
```

```python
#testing the functions
nR,nC = 8,8

A = np.zeros( [nR,nC], dtype=int)

glider(A, 4,4)

plt.subplot(1,3,1)
plt.imshow(A)

plt.subplot(1,3,2)
plt.imshow(checkNeighbors(A))

plt.subplot(1,3,3)
plt.imshow(nextGen(A, checkNeighbors(A)))
```

```python
#testing the functions in a movie
nR = 31
nC = 31
nT = 101
    
MOVIE = np.zeros( [nR,nC,nT], dtype=int) #make array for whole screen
    
glider(MOVIE[:,:,0], 4, 4)

for timestamp in range(nT - 1):
    neighbors = checkNeighbors(MOVIE[:,:,timestamp])
    nextFrame = nextGen(MOVIE[:,:,timestamp], neighbors)
    MOVIE[:,:,timestamp+1] = nextFrame

plt.imshow(MOVIE[:,:,10])
```

```python
golMovie = ani.FuncAnimation(fig    = golFig, 
                             func   = golAnimate, 
                             frames = nT,
                             fargs  = (MOVIE,golPlot)
                             );

HTML(golMovie.to_jshtml())
```

```python
#This is the producer    
def golProducer(M, Q):
    
    nT = M.shape[2]
    
    for timestamp in np.arange(nT):
        flag = int(rng.random()*100) % 10
        locationR = int(rng.random()*100) % nR
        locationC = int(rng.random()*100) % nC
        
        if flag == 1:
            glider(M[:,:,timestamp], locationR,locationC)
            
        neighbors = checkNeighbors(M[:,:,timestamp])
        nextFrame = nextGen(M[:,:,timestamp], neighbors)
        try:
            M[:,:,timestamp+1] = nextFrame
        except:
            M = M
        
        Q.put((timestamp, flag))
        
            
#The producer inserts a specific shape at random locations in its field
#for simplicity, the producer uses only the "glider" shape (see below)
#add the glider to the field only at 10% of the time steps selected randomly
#and puts into a queue at every generation its index and a flag * the boolean flag specifies if at least 25 cells are alive in its field at that time
```

```python
#this is the consumer
def golConsumer(M,Q):
    
    nT = M.shape[2]
    
    while True:
        
        timestamp, flag = Q.get()
        if timestamp == nT - 1:
            break
            
        neighbors = checkNeighbors(M[:,:,timestamp])
        nextFrame = nextGen(M[:,:,timestamp], neighbors)
        M[:,:,timestamp+1] = nextFrame
        
        
        locationR = int(rng.random()*100) % nR
        locationC = int(rng.random()*100) % nC
        
        if flag == 1:
            glider(M[:,:,timestamp], locationR,locationC)
        Q.task_done()
        
        
            
#The consumer receives the information from the queue and
#inserts a glider into the field at a random position every time the received flag is True.
#Once a glider is inserted, the consumer does not insert another one for at least 5 generations
```

```python
#this is the queue and threading cell
Q = queue.Queue()

AMOVIE = np.zeros( [nR,nC,nT], dtype=int)
AMOVIE[0,0,0] = 1
BMOVIE = np.zeros( [nR, nC, nT], dtype=int)
BMOVIE[0,0,0] = 1

P = threading.Thread(target = golProducer, args= (AMOVIE, Q))
C = threading.Thread(target = golConsumer, args=(BMOVIE, Q))

P.start()
C.start()

Q.join()
```

```python
#this plot is the producer animation
golFig = plt.figure()

golPlotA = plt.imshow(AMOVIE[:,:,0])

golMovieA = ani.FuncAnimation(fig    = golFig, 
                             func   = golAnimate, 
                             frames = nT,
                             fargs  = (AMOVIE,golPlotA)
                             );

HTML(golMovieA.to_jshtml())       
```

```python
#This plot is the consumer animation
golFig = plt.figure()

golPlotB = plt.imshow(BMOVIE[:,:,0])

golMovieB = ani.FuncAnimation(fig    = golFig, 
                             func   = golAnimate, 
                             frames = nT,
                             fargs  = (BMOVIE,golPlotB)
                             );

HTML(golMovieB.to_jshtml())  
```

```python
#This plot is meant to make two animations in one function
def animateTwo(i,AMOVIE, BMOVIE, ax1, ax2):
    ax1.set_data(AMOVIE[:,:,i])
    plt.axis('off')
    ax2.set_data(BMOVIE[:,:,i])
    plt.axis('off')
```

```python
#The above function should be used to make a double animation in this cell

fig, (axl, axr) = plt.subplots(
    ncols=2,
    sharey=True,
    figsize=(6, 6))

timeline = np.arange(nT)

golPlot1 = plt.imshow(AMOVIE[:,:,0]);
plt.axis('off') # make a plot
golPlot2 = plt.imshow(BMOVIE[:,:,0]);
plt.axis('off');


golMovie = ani.FuncAnimation(fig    = fig, 
                             func   = animateTwo, 
                             frames = nT,
                             fargs  = (AMOVIE, BMOVIE, golPlot1, golPlot2)
                             );

HTML(golMovie.to_jshtml()) 

```
