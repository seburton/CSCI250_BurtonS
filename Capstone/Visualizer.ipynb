# This is a visualizer for audio data. Load the file and see data plot below.

```python
####################### Add the name of the file you wish to view below. ###############################################################################
filename = "newfile1.npy"

data = np.load(filename, allow_pickle=True)

duration = data[0].duration

time = np.arange(0, duration * len(data), duration)

frequencies = []
for i in range(len(data)):
    frequencies.append(data[i].pitch)

plt.plot(time, frequencies)
plt.title("file: " + filename)
plt.xlabel("time (s)")
plt.ylabel("frequency (Hz)")
```
